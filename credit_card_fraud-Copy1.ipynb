```python
from imblearn.over_sampling import SMOTE
```

```python
df_3 = df_2.copy()
```

```python
y_train.value_counts()
```

```python
df_3_smote,df_3_target_smote = SMOTE().fit_resample(df_3,y_train)
```

```python
df_3_target_smote.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=25, max_features='sqrt', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=2, min_samples_split=50,
            min_weight_fraction_leaf=0.0, n_estimators=700, n_jobs=None,
            oob_score=False, random_state=None, verbose=0,
            warm_start=False)
```

```python
model.fit(df_3_smote,df_3_target_smote)
```

```python
model.score(df_3_smote,df_3_target_smote)
```

```python
x_test =pd.get_dummies(x_test,columns = ['GENDER','CAR','REALITY','INCOME_TYPE','FAMILY_TYPE','HOUSE_TYPE','EDUCATION_TYPE'],drop_first = True)
```

```python
x_test.drop('NO_OF_CHILD',axis=1,inplace=True)
```

```python
x_test,y_test= SMOTE().fit_resample(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_pred,y_test))
```
